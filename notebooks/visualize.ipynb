# Twinkle + dsDNA Visualization

This notebook is for showing the prepared local environment, not a finished simulation trajectory.

Recommended show order:
1. merged Twinkle + dsDNA complex,
2. solvated local environment,
3. final HPC handoff bundle summary.

```python
from pathlib import Path
import json

import MDAnalysis as mda
import nglview as nv

ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
CG_DIR = ROOT / "cg"
HANDOFF_DIR = CG_DIR / "handoff_m2_mg_proxy"

complex_pdb = CG_DIR / "complex_cg.pdb"
inspection_pdb = CG_DIR / "solvated_inspection_system.pdb"
inspection_json = CG_DIR / "solvated_inspection_system.json"
handoff_pdb = HANDOFF_DIR / "solvated_m2_mg_proxy.pdb"

for path in [complex_pdb, inspection_pdb, inspection_json, handoff_pdb]:
    print(path.name, "OK" if path.exists() else "MISSING")
```

```python
u_complex = mda.Universe(str(complex_pdb))
view_complex = nv.show_mdanalysis(u_complex)
view_complex.clear_representations()
view_complex.add_representation("ball+stick", selection="protein", color="skyblue")
view_complex.add_representation("ball+stick", selection="nucleic or resname DA* DC* DG* DT*", color="orange")
view_complex.add_representation("spacefill", selection="resname NA CL MG MG2", color="red", opacity=0.6)
view_complex.center()
view_complex
```

Use the first viewer to show that the dsDNA passes through the Twinkle ring channel.

```python
metadata = json.loads(inspection_json.read_text())
metadata["counts"]
```

```python
u_env = mda.Universe(str(inspection_pdb))
view_env = nv.show_mdanalysis(u_env)
view_env.clear_representations()
view_env.add_representation("cartoon", selection="protein", color="skyblue")
view_env.add_representation("licorice", selection="nucleic or resname DA* DC* DG* DT*", color="orange")
view_env.add_representation("line", selection="resname W", color="lightgray", opacity=0.15)
view_env.add_representation("spacefill", selection="resname NA CL MG", opacity=0.5)
view_env.center()
view_env
```

Use the second viewer to show that the local environment exists: protein, DNA, water, and ion population are all present.
